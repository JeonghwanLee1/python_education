<b>인공신경망(Neural Network)</b><br>
- 인공 뉴런들을 모아 다양한 입력, 출력 형태를 다룰 수 있는 신경망이다.
- 가장 단순한 형태의 신경망부터 딥 러닝이라고 불리는 분야에서 활용되는 매우 복잡한 구조를 가진 신경망까지 그 형태가 매우 다양하다.

<b>CNN(Convolution Neural Network)이란 무엇일까?</b><br>
- 딥 러닝 네트워크에서 특별한 연산과 계층이 추가된 네트워크이다.
- Convolution, Pooling 등의 연산과 해당 연산을 수행하는 계층이 추가된다.
- 이미지 인식 등의 분야에서 많이 활용된다.

<b>합성곱(Convolution) 이란?</b><br>
- 특수한 필터를 이용해 기존의 데이터를 필터링 하는 연산이다.
- 합성곱의 결과는 데이터 행렬과  필터 행렬을 곱한 결과가 된다.
- 합성곱이 완료된 데이터는 일반적으로 기존의 데이터보다 크기가 작아진다.

<b>Pooling 이란?</b><br>
- 데이터의 크기를 줄이는 작업(Subsampling)이다.
- 최대값을 기준으로 크기를 줄이는 Max Pooling, 평균을 기준으로 줄이는 Mean Pooling 등이 있지만, 주로 Max Pooling을 사용한다.
- 합성곱이 완료된 데이터는 일반적으로 기존의 데이터보다 크기가 작아진다.

<b>Inception-v3 모델 불러오기 </b><br>
- 이미지를 활용한 학습에는 시간과 자원이 많이 소요되므로, 해당 실습에서는 미리 학습된 구글의 Inception-v3 모델을 이용한다.
- 주어진 소스 코드를 활용해 Inception-v3 모델을 불러오고, 모델의 성능을 확인해 보자.

```python
from tensorflow.keras.applications.inception_v3 import InceptionV3,     decode_predictions
import tensorflow as tf
inceptionv3 = InceptionV3(include_top=True, weights='imagenet', input_tensor=None,  input_shape=None, pooling=None, classes=1000)
```

<b>모델 테스트</b><br>
- 임의의 이미지 파일을 이용해 모델을 테스트 해보자.


```python
from matplotlib import pyplot as plt
from PIL import Image
file_path = 'elephant.jfif'
image = Image.open(file_path)
#pyplot을 이용한 이미지 출력
plt.figure(figsize=(11,11))
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
import numpy as np
#모델이 지원하는 사이즈인 299*299로 맞춤
width = 299 
height = 299
image = image.resize((width, height))
image = np.array(image) 
x_test = [image]
x_test =  np.array(x_test)
x_test = x_test / 255

y_predict = inceptionv3.predict(x_test)
label = decode_predictions(y_predict)
print("predict : ",label[0][0][1])
print("probability : ",label[0][0][2]*100,"%")
```
