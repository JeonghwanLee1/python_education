#### plt와 show
- matplotlib 중 우리가 필요로 하는 기능을 사용하기 위해, matplotlib 라이브러리 중 pyplot을 import 한다.

#### [import 예제]

```python
from matplotlib import pyplot as plt     #alias 사용
```

- pyplot.plot을 통해 데이터를 이용한 plot을 만들고 pyplot.show()를 통해 만들어진 plot을 기초적인 그래프로 시각화한다.

#### [plot과 show]

```python
from matplotlib import pyplot as plt
plt.plot([1,2,3],[100,300,500])
plt.show()
```

- 이 때, plot은 첫 번째 들어온 리스트를 x축, 두 번째 들어온 리스트를 y축의 값으로 설정하며 특별한 지시가 없다면 해당 리스트들의 수치를 고려해 가장 이상적인 축의 단위값을 설정한다.

#### title과 xlabel, ylabel

<div style = "border : 1px solid green; padding : 5px; width = 50px;">
pyplot.title(title) <br>
pyplot.xlabel(name) , pyplot.ylabel(name) <br>
<div style = "color : gray;">
title : 제목 <br>
name : 레이블 이름 
</div></div>

- pyplot.title(x), pyplot.xlabel(x), pyplot.ylabel(x) 를 통해 차트의 제목과 축 레이블 이름을 지정할 수 있다.

#### [title, xlabel, ylabel]

```python
from matplotlib import pyplot as plt
plt.plot([1,2,3],[100,300,500])
plt.title("velocity per time")
plt.xlabel("time")
plt.ylabel("velocity")
plt.show()
```

- 기본적으로 제목은 차트의 위쪽에, 나머지 축의 레이블 이름은 각각의 축 의 가운데에 위치하게 된다. 

#### xlim, ylime

<div style = "border : 1px solid green; padding : 5px; width = 50px;">
pyplot.xlim(min, max), pyplot.ylim(min, max)<br>
<div style = "color : gray;">
xlim, ylim : 최소, 최대값
</div></div>

- x축은 pyplot.xlim, y축은 pyplot.ylim로 구간의 설정이 가능하다.

#### [xlim, ylim]

```python
from matplotlib import pyplot as plt
plt.plot([1, 2, 3], [100, 200, 300])
plt.title("velocity per time")

#x축 구간 지정
plt.xlim(1.5, 2.5)

#y축 구간 지정
plt.ylim(0, 1000)

plt.xlabel("time")
plt.ylabel("velocity")
plt.show()
```

#### xscale과 yscale
<div style = "border : 1px solid green; padding : 5px; width = 50px;">
pyplot.xscale(scale), pyplot.yscale(scale) <br>
<div style = "color : gray;">
scale : 스케일
</div></div>

- pyplot.xscale, pyplot.yscale을 이용하여 x축과 y축의 스케일을 조절 할 수 있다.

#### [xscale과 yscale]

```python
plt.plot([1,2,3],[100,1000,10000])
plt.title("velocity per time")
plt.xlabel("time")
plt.xscale("log")     #log스케일
plt.ylabel("velocity")
plt.show()
```

- 설정 가능한 스케일의 종류는 log 외에 symlog, linear 등이 있다. 더 알고 싶다면 이 부분에 대해 조사해 보자.

#### 어노테이션
<div style = "border : 1px solid green; padding : 5px; width = 50px;">
pyplot.annotate(str, pos, textpost, props) <br>
<div style = "color : gray;"><br>
   
str : 문자열, pos : 좌표<br>
textpost : 문자열 좌표, props : 화살표 속성
</div></div>

- 어노테이션을 이용해 화살표와 그 화살표에 문자열을 출력할 수 있다.  
- plt.annotate('문자열',좌표, 출력될 위치, 화살표 속성) 함수를 이용한다.

#### [어노테이션의 사용]

```python
plt.plot([1,2,3],[100,200,300])
plt.title("velocity per time")
plt.xlabel("time")
plt.ylabel("velocity")
plt.annotate('threshold', xy=(2,200), xytext=(2,300), arrowprops={'color':'blue'})
plt.show()
```

#### 서브플롯

<div style = "border : 1px solid green; padding : 5px; width = 50px;">
pyplot.subplot(row, col, pos) <br>
<div style = "color : gray;">
row : 행, col : 열, pos : 위치
</div></div>

- 한 화면에 여러개의 그래프를 그리고 싶을 경우 서브플롯을 사용한다.
- subplot(row,col,pos) 로 여러 개의 그래프를 담을 표를 정의하고, 해당 플롯의 위치를 정의한다.

#### [서브플롯의 사용]

```python
plt.subplot(2,1,1)
plt.plot([1,2,3],[100,300,500])
plt.subplot(2,1,2)
plt.plot([2,3,4],[200,300,400])
plt.show()
```

- subplot(row,col,pos)는 각각 행, 열, 위치를 나타낸다. 예를 들어 subplot(2,2,1) 이라면, 2행 2열로 이루어진 공간에서 첫번째 위치를 나타낸다

#### [서브플롯의 사용]

```python
plt.subplot(2,2,1)     #2행 2열 공간 중 첫번째
plt.plot([1,2,3],[100,300,500])
plt.subplot(2,2,2)     #2행 2열 공간 중 두번째
plt.plot([2,3,4],[200,300,400])
plt.subplot(2,2,3)     #2행 2열 공간 중 세번째
plt.plot([2,3,4],[10,30,40])
plt.subplot(2,2,4)     #2행 2열 공간 중 네번째
plt.plot([2,3,4],[20,30,40])
plt.show()
```

#### 선 옵션 변경
- pyplot.plot() 함수 내부의 옵션으로 선의 스타일을 변경할 수 있다.  

#### [선 옵션 변경하기]

```python
plt.plot([1,2,3],[100,300,500],'-')     #실선(선 옵션)
plt.plot([2,3,4],[200,300,400],'--')     #점선(선 옵션)
plt.plot([1,2,3],[100,150,200],'o')     #점(마커 옵션)
plt.plot([1,2,3],[300,400,500],'-o')     #점과 실선(마커 옵션과 선 옵션)

plt.title("velocity per time")
plt.xlabel("time")
plt.ylabel("velocity")
plt.legend(["car1","car2","car3","car4"])
plt.show()
```

### 2. 다음과 같은 데이터셋을 이용해 그래프를 그려 보자.
아래와 같은 데이터셋을 활용하여 문제에서 요구하는 그래프를 그려 보자.  
- scatter 형 그래프 
- x의 label은 height 
- x의 범위는 140~200
- y의 label은 weight 
- 평균을 나타내는 점에 어노테이션 표시

```python
data_x = [170,180,160,150,166,146,177,160,184,171]
data_y = [60,77,71,45,66,44,70,66,80,70]
x_mean = sum(data_x)/len(data_x)
y_mean = sum(data_y)/len(data_y)
plt.scatter(data_x,data_y,color='pink')
plt.xlabel("height")
plt.xlim(140,200)
plt.ylabel("weight")
plt.annotate('mean', xy=(x_mean,y_mean),
             xytext=(x_mean,y_mean+20), arrowprops={'color':'blue'})
```
