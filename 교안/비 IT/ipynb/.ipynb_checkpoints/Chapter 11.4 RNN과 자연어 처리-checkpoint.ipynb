<b>RNN의 문제점</b>
- 기준 정보와 그 정보를 사용하는 지점의 거리가 멀 경우, 학습 능력이 저하되는 문제가 발생한다(Vanishing Gradient).
- 경사 하강법(Gradient Descent)의 고질적인 문제점이다.
- 기존 RNN에서 이 문제점을 극복한 모델이 바로 LSTM이다.

<b>LSTM</b>
- 기존 RNN의 hidden state에 cell-state를 추가한 모델이다.
- Long Short-Term Memory 의 사용으로 기존 RNN의 긴 의존기간의 문제(Long-term dependencies)를 해결했다.
- RNN의 vanishing gradient 문제를 해결했다.

<b> 셰익스피어 소설을 학습하는 RNN 예제 구현 절차</b>
- 학습 환경 설정
- 데이터 불러오기
- 2 Hidden Layer를 포함한 RNN(LSTM) 구현
- 학습을 진행하며 각 Epoch마다 변화 확인
- 학습 완료 후, 샘플링을 통해 새 소설 창작

<b>학습 환경 설정</b>
- 학습 환경을 설정한다
- 한번에 학습할 데이터 수는 50개
- 한번에 학습할 문장의 길이는 50
- Hidden layer의 노드 수는 128
- 학습 속도는 0.002
- 전체 데이터셋을 2번 학습

```python
import tensorflow.compat.v1 as tf
import numpy as np
tf.disable_eager_execution()
from utils import TextLoader
import rnn
data_dir = 'data/tinyshakespeare'
batch_size = 50
seq_length = 50 
hidden_size = 128 
learning_rate = 0.002
num_epochs = 2
num_hidden_layers = 2
grad_clip = 5 
```

<b>데이터 불러오기</b>
- 기존에 저장된 셰익스피어 소설 데이터를 구현된 Textloader 모듈을 통해 불러온다.

```python
data_loader = TextLoader(data_dir, batch_size, seq_length)
# 학습데이터에 포함된 모든 단어들을 나타내는 변수인 chars와 chars에 id를 부여.
chars = data_loader.chars 
vocab = data_loader.vocab
vocab_size = data_loader.vocab_size # 총 단어수
```

<b> RNN 모델 구현</b>
- 인풋의 형태를 정의한다.
- 활성 함수를 정의한다.
- 최종 출력값과 손실 함수를 정의한다.
- 옵티마이저를 설정한다.
- 이후 학습에 필요한 모델 구조를 설정한다.

```python
#인풋 데이터 placeholder
input_data = tf.placeholder(tf.int32, shape=[None, None])
#타겟 데이터 placeholder
target_data = tf.placeholder(tf.int32, shape=[None, None]) 
state_batch_size = tf.placeholder(tf.int32, shape=[])

#softmax 함수 정의
softmax_w = tf.Variable(tf.random_normal(shape=[hidden_size,vocab_size]),            dtype=tf.float32)
softmax_b = tf.Variable(tf.random_normal(shape=[vocab_size]),   dtype=tf.float32)

#hidden layer(LSTM) 선언
cells = []
for _ in range(0, num_hidden_layers):
    cell = tf.nn.rnn_cell.BasicLSTMCell(hidden_size)
    cells.append(cell)

#종합하여 RNN 선언
cell = rnn.MultiRNNCell(cells, state_is_tuple=True)

# 인풋데이터를 변환하기 위한 과정
embedding = tf.Variable(tf.random_normal(shape=[vocab_size, hidden_size]), dtype=tf.float32)
inputs = tf.nn.embedding_lookup(embedding, input_data)

# 초기 state 값을 0으로 초기화
initial_state = cell.zero_state(state_batch_size, tf.float32)

# 학습을 위한 tf.nn.dynamic_rnn을 선언
outputs, final_state = tf.nn.dynamic_rnn(cell, inputs, initial_state=initial_state, dtype=tf.float32)
#output 모양 설정
output = tf.reshape(outputs, [-1, hidden_size])

# 최종 출력값을 설정
logits = tf.matmul(output, softmax_w) + softmax_b
probs = tf.nn.softmax(logits)

#Cross Entropy 손실 함수 정의
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=logits, labels=target_data))

#옵티마이저를 선언
tvars = tf.trainable_variables()
grads, _ = tf.clip_by_global_norm(tf.gradients(loss, tvars), grad_clip)
optimizer = tf.train.AdamOptimizer(learning_rate)
train_step = optimizer.apply_gradients(zip(grads, tvars))
```

<b> 모델 설정 및 학습 진행</b>
- 인풋의 형태를 정의한다.
- 활성 함수를 정의한다.
- 최종 출력값과 손실 함수를 정의한다.
- 옵티마이저를 설정한다.
- 이후 학습에 필요한 모델 구조를 설정한다.

```python
with tf.Session() as sess:
    # 변수들에 초기값 할당
    sess.run(tf.global_variables_initializer())
    
    for e in range(num_epochs):
        data_loader.reset_batch_pointer()
        # 초기 상태값 지정
        state = sess.run(initial_state, feed_dict={state_batch_size : batch_size})

        for b in range(data_loader.num_batches):
            x, y = data_loader.next_batch()
            y = tf.one_hot(y, vocab_size)           
            y = tf.reshape(y, [-1, vocab_size])      
            y = y.eval()

            feed_dict = {input_data : x, target_data: y, state_batch_size : batch_size}
            for i, (c, h) in enumerate(initial_state):
                feed_dict[c] = state[i].c
                feed_dict[h] = state[i].h

            _, loss_print, state = sess.run([train_step, loss, final_state], feed_dict=feed_dict)

            print("{}/{}, epoch: {}, 손실함수(loss): {}".format(
                          e * data_loader.num_batches + b,
                          num_epochs * data_loader.num_batches,
                          (e+1), 
                          loss_print))    

    # 샘플링 시작
    print("문서 생성 중")
    num_sampling = 4000  # 생성할 글자(Character)의 개수 
    prime = u' '         # 시작 글자 ' '(공백)으로 지정
    sampling_type = 1 
    state = sess.run(cell.zero_state(1, tf.float32)) 

    #이전 글자를 보고 다음 글자를 고르는 함수
    def weighted_pick(weights):
        t = np.cumsum(weights)
        s = np.sum(weights)
        return(int(np.searchsorted(t, np.random.rand(1)*s)))

    ret = prime      
    char = prime      
    for n in range(num_sampling):
        x = np.zeros((1, 1))
        x[0, 0] = vocab[char]

        # RNN을 한스텝 실행
        feed_dict = {input_data: x, state_batch_size : 1, initial_state: state}
        [probs_result, state] = sess.run([probs, final_state], feed_dict=feed_dict)         

        # 불필요한 차원 제거
        p = np.squeeze(probs_result)

        # sampling_type : 0 -> 다음 글자를 예측할때 항상 argmax를 사용
        # sampling_type : 1(defualt) -> 다음 글자를 예측할때 항상 random sampling을 사용
        # sampling_type : 2 -> 다음 글자를 예측할때 이전 글자가 ' '(공백)이면 random sampling, 그렇지 않을 경우 argmax를 사용
        if sampling_type == 0:
            sample = np.argmax(p)
        elif sampling_type == 2:
            if char == ' ':
                sample = weighted_pick(p)
            else:
                sample = np.argmax(p)
        else:
            sample = weighted_pick(p)

        pred = chars[sample]
        ret += pred     # 샘플링 결과에 현재 스텝에서 예측한 글자를 추가
        char = pred     # 예측한 글자를 다음 RNN의 인풋으로 사용
```

<b> 결과 확인</b>
- 학습한 모델이 작성한 결과를 확인해보자

```python
print(ret)
```
