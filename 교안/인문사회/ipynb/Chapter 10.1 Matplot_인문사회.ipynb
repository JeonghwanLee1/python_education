### 10.1 Matplotlib
#### Matplotlib 소개
- Matplot이란 다양한 데이터를 많은 방법으로 도식화 할 수 있는 기능을 제공하는 python의 라이브러리이다.
- 데이터를 차트나 그래프의 형태의 시각화 자료로 보여주기 때문에 mathworks에서 개발한 매트랩과 유사한 형태를 제공한다.
- 라인 플롯, 바 차트, 파이 차트, 히스토그램, 박스 플롯, 스카터 플롯 등을 비롯한 다양한 종류의 차트와 스타일을 지원하며 이와 같은 자료들은 Matplotlib.org 갤러리에서 샘플 차트를 볼 수 있다.

#### [plot과 show]

```python
    #alias 사용 import
    from matplotlib import pyplot as plt

    #데이터의 값 : (100, 900), (225, 150), (900, 1500)
    plt.plot([100, 225, 900], [900, 150, 1500])

    #차트의 시각화
    plt.show()
```

- pyplot.title(x), pyplot.xlabel(x), pyplot.ylabel(x) 를 통해 차트의 제목과 축 레이블 이름을 지정할 수 있다.

#### xlim, ylime
- 그래프의 최소 구간과 최대 구간을 설정하기 위하여 x축은 pyplot.xlim, y축은 pyplot.ylim를 사용한다.

<div style = "color : gray;">

#### [그래프의 최소 구간 및 최대 구간 설정]

```python
    plt.plot([100, 500, 1000], [60, 120, 180])
    plt.title("meter per second")
    plt.xlabel("meter")
    plt.ylabel("second")

    #x축 구간 지정 : (최소값, 최대값)
    plt.xlim(0, 2000)
    #y축 구간 지정 : (최소값, 최대값)
    plt.ylim(0, 600)

    plt.show()
```

#### subplot
- 한 화면에 여러 개의 그래프를 그릴 때 subplot을 사용한다.
- 이때 매개변수를 통해 표의 크기와 위치의 설정이 가능하다.

<div style = "border : 1px solid green; padding : 5px; width = 50px;">
pyplot.subplot(row, col, pos) <br>
<div style = "color : gray;">
row : 행, col : 열, pos : 위치
</div></div>

#### [subplot을 사용하여 여러 개의 그래프 추가하기(1)]

```python
    #3행 1열로 이루어진 표의 첫번째 위치
    plt.subplot(3, 1, 1)
    plt.plot([1, 2, 3], [100, 200, 300])

    #3행 1열로 이루어진 표의 두번째 위치
    plt.subplot(3, 1, 2)
    plt.plot([1, 2, 3], [200, 300, 400])
    plt.show()

    #3행 1열로 이루어진 표의 세번째 위치
    plt.subplot(3, 1, 3)
    plt.plot([1, 2, 3], [200, 300, 400])

    plt.show()
```

#### [subplot을 사용하여 여러 개의 그래프 추가하기(2)]

```python
    #4행 1열로 이루어진 표의 첫번째 위치
    plt.subplot(4, 1, 1)
    plt.plot([0, 0.5, 1], [10, 20, 30])

    #4행 2열로 이루어진 표의 첫번째 위치
    plt.subplot(4, 1, 2)
    plt.plot([0, 0.5, 1], [0, 1, 2])
 
    #4행 3열로 이루어진 표의 첫번째 위치
    plt.subplot(4, 1, 3)
    plt.plot([0, 0.5, 1], [5, 1.5, 5])
 
    #4행 4열로 이루어진 표의 첫번째 위치
    plt.subplot(4, 1, 4)
    plt.plot([0, 0.5, 1], [1, 2, 3])

    plt.show()
```

#### 선 옵션 변경
- pyplot.plot() 함수 내부의 옵션으로 그래프의 스타일을 변경할 수 있다.
- pyplot에서 사용 가능한 마커의 종류는 다음과 같다. 
<br><br>
<center><표 10.1-3>마커 옵션</center>
<img src="https://user-images.githubusercontent.com/37530732/75514283-7b369880-5a3a-11ea-9120-849bee042028.png" style="width: 80%;"/><br>
    
#### [그래프 스타일 변경(1)]

```python
    #실선(선 옵션)
    plt.plot([0, 0.5, 1], [10, 20, 30], '-')

    #점선(선 옵션)
    plt.plot([0, 0.5, 1], [0, 1, 2], '--')

    #점(마커 옵션)
    plt.plot([0, 0.5, 1], [5, 1.5, 5], 'o')

    plt.title("meter per second")
    plt.xlabel("meter")
    plt.ylabel("second")
    plt.legend(["car1", "car2", "car3"])

    plt.show()
```

#### [그래프 스타일 변경(2)]

```python
    #실선(선 옵션)
    plt.plot([0, 0.5, 1], [10, 20, 30], '-')

    #점선 + x모양 마커
    plt.plot([0, 0.5, 1], [0, 1, 2], '--')

    #다이아몬드 마커
    plt.plot([0, 0.5, 1], [5, 1.5, 5], 'o')

    plt.title("meter per second")
    plt.xlabel("meter")
    plt.ylabel("second")
    plt.legend(["car1", "car2", "car3"])

    plt.show()
```

#### Scatter형 그래프
- pyplot.scatter 함수를 이용하여 산포 그래프를 작성할 수 있다.
- 산포 그래프란 다양한 마커 사이즈와 색상을 이용하여 만든 plot이다.

<div style = "border : 1px solid green; padding : 5px; width = 50px;">
pyplot.scatter(x, y, s= None, c=None)<br>
<div style = "color : gray;">
x, y : 데이터
</div></div>

- 위의 형식에서 s는 선택적으로 생략 가능하며 마커의 크기를 설정한다.
- c는 plot내에서 x,y와 같이 요수의 수를 가지는 반복적인 자료형을 입력할 경우 각 마커마다 다른 색상을 설정할 때 지정하며 생략 가능하다.

#### [scatter를 이용한 산포 그래프]

```python
    #개별데이터
    plt.scatter([165, 180, 171, 173, 160, 159, 183, 182],
             [53, 77, 70, 80, 50, 44, 83, 90])

    plt.title("weight per height")
    plt.xlabel("height")
    plt.ylabel("weight")

    plt.show()
```
